```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
x_train
```

```python
i= 0 

print(x_train[i],y_train[i])
```

```python
plt.scatter(x_train,y_train ,c='r',marker ="^")
plt.title("house price pred")

plt.xlabel("house size in sqft")
plt.ylabel("house prize in 1000")
plt.legend()
plt.show()
```

```python
w=200
b=100
```

```python
def com_model_op(x,w,b): 
    m = x.shape
    f_wb = np.zeros(m)
    
    for i in range(len(x)):
        f_wb[i] = w*x[i] + b 
        
    return f_wb
```

```python
tmp_f_wb = com_model_op(x_train,w,b)

plt.plot(x_train,tmp_f_wb,c="y",label="our prediction ")

plt.scatter(x_train,y_train,c="r",label="actual value")

plt.title("HPP")
plt.xlabel("house size")
plt.ylabel("house prize ")
plt.legend()
plt.show()
```
